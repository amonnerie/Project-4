# Clean Up Data

```python
import pandas as pd
import datetime as dt
```

```python
energy_df = pd.read_csv('Clean-up/energy_dataset_clean_excel.csv', encoding='UTF-8')
weather_df = pd.read_csv('Clean-up/weather_features_clean_excel.csv', encoding='UTF-8')
energy_df.head()
```

```python
weather_df.head()
```

```python
datetimezone = energy_df['time'].astype("string")
datelist = []
timelist = []

for datetimerow in datetimezone:
    date, time = datetimerow.split(" ", 1)
    time, timezone = time.split("+", 1)
    datelist.append(date)
    timelist.append(time)
```

```python
datetimezone = weather_df['dt_iso'].astype("string")
datelist2 = []
timelist2 = []

for datetimerow in datetimezone:
    date, time = datetimerow.split(" ", 1)
    time, timezone = time.split("+", 1)
    datelist2.append(date)
    timelist2.append(time)
```

```python
energy_df['date'] = datelist
energy_df['time_24h'] = timelist
weather_df['date'] = datelist2
weather_df['time_24h'] = timelist2 
```

```python
energy_df.head()
```

```python
weather_df.head()
```

```python
energy_df = energy_df.dropna()
energy_df.isna().any(axis=1)
```

```python
weather_df = weather_df.dropna()
weather_df.isna().any(axis=1)
```

```python
energy_df.to_csv("Clean-up/energy_dataset_clean_python.csv", index=False, encoding='utf-8')
```

```python
weather_df.to_csv("Clean-up/weather_features_clean_python.csv", index=False, encoding='utf-8')
```
