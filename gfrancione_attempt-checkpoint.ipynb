```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named energy_weather_df with the relevant data

# Convert the 'time' column to datetime with utc=True
energy_weather_df['time'] = pd.to_datetime(energy_weather_df['time'], utc=True)

# Extract the year and month from the 'time' column
energy_weather_df['Year'] = energy_weather_df['time'].dt.year
energy_weather_df['Month'] = energy_weather_df['time'].dt.month

# Group the data by year and month
grouped_data = energy_weather_df.groupby(['Year', 'Month'])['total load actual'].sum().reset_index()

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Iterate through each year and plot the data with different colors
for year, data in grouped_data.groupby('Year'):
    ax.plot(data['Month'], data['total load actual'], marker='o', label=f'Year {year}')

# Set plot properties
plt.title('Total Energy Consumption by Year')
plt.xlabel('Month')
plt.ylabel('Total Energy Consumption')
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Review/Create a histogram of energy prices
plt.figure(figsize=(10, 6))
plt.hist(energy_weather_df['price actual'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Energy Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)

# Optionally, add a vertical line for the mean or other summary statistics
mean_price = energy_weather_df['price actual'].mean()
plt.axvline(mean_price, color='red', linestyle='dashed', linewidth=2, label=f'Mean Price: {mean_price:.2f}')
plt.legend()

plt.show()
```

```python

# Filter the DataFrame to only include data related to solar generation
solar_generation_df = energy_weather_df[['time', 'generation solar']]

# Convert the 'time' column to a datetime object if it's not already
solar_generation_df['time'] = pd.to_datetime(solar_generation_df['time'])

# Extract the year and month from the 'time' column
solar_generation_df['Year'] = solar_generation_df['time'].dt.year
solar_generation_df['Month'] = solar_generation_df['time'].dt.month

# Group the data by year and month and sum the solar generation
#What else could this be 
monthly_solar_generation = solar_generation_df.groupby(['Year', 'Month'])['generation solar'].avg()

# Reset the index to convert the result to a DataFrame
monthly_solar_generation = monthly_solar_generation.reset_index()

# Define a dictionary to map month numbers to month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Map month numbers to month names for better labeling
monthly_solar_generation['Month'] = monthly_solar_generation['Month'].map(month_names)

# Create a bar plot to show monthly solar generation
plt.figure(figsize=(12, 6))
plt.bar(monthly_solar_generation.index, monthly_solar_generation['generation solar'])
plt.title('Monthly Solar Generation')
plt.xlabel('Month')
plt.ylabel('Solar Generation (MW)')
plt.xticks(monthly_solar_generation.index, monthly_solar_generation['Month'], rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named energy_weather_df with the relevant data

# Filter the DataFrame to include the 'time' and 'price actual' columns
price_df = energy_weather_df[['time', 'price actual']]

# Convert the 'time' column to a datetime object if it's not already
price_df['time'] = pd.to_datetime(price_df['time'])

# Extract the year and month from the 'time' column
price_df['Year'] = price_df['time'].dt.year
price_df['Month'] = price_df['time'].dt.month

# Group the data by year and month and calculate the average price for each month
monthly_avg_price = price_df.groupby(['Year', 'Month'])['price actual'].mean()

# Reset the index to convert the result to a DataFrame
monthly_avg_price = monthly_avg_price.reset_index()

# Create a line plot to show the average energy price for each month
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg_price.index, monthly_avg_price['price actual'], marker='o', linestyle='-', color='b')
plt.title('Average Monthly Energy Price')
plt.xlabel('Month')
plt.ylabel('Price (EUR/MWh)')
plt.xticks(monthly_avg_price.index, monthly_avg_price['Month'])
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Assuming you have a DataFrame named energy_weather_df with the relevant data

# Filter the DataFrame to include the 'time' and 'price actual' columns
price_df = energy_weather_df[['time', 'price actual']]

# Convert the 'time' column to a datetime object if it's not already
price_df['time'] = pd.to_datetime(price_df['time'])

# Extract the year and month from the 'time' column
price_df['Year'] = price_df['time'].dt.year
price_df['Month'] = price_df['time'].dt.month

# Define a dictionary to map month numbers to full month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Map the month numbers to full month names
price_df['Month'] = price_df['Month'].map(month_names)

# Group the data by year and month and calculate the average price for each month
monthly_avg_price = price_df.groupby(['Year', 'Month'])['price actual'].mean()

# Reset the index to convert the result to a DataFrame
monthly_avg_price = monthly_avg_price.reset_index()

# Create a new column combining Year and Month for better display
monthly_avg_price['Month_Year'] = monthly_avg_price.apply(lambda row: f"{row['Month']}-{row['Year']}", axis=1)

# Display the average price per month with the Month-Year label and full month name
print(monthly_avg_price[['Month_Year', 'price actual']])
```
